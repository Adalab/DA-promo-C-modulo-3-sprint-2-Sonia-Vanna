```python
#Cargamos el dataset CUIDADO ESTA CON LA DIRECCION DE DRIVE
df=pd.read_csv('/content/drive/My Drive/Colab Notebooks/ADALAB/pair/ufo_fullset.csv')
df.head(3)
```

```python
df.columns
```

```python
#Eliminamos las comillas en los nombres de las columnas
columnas_nuevas = {col :  col.lower().replace('"', "") for col in df.columns}
df.rename(columns = columnas_nuevas, inplace = True)
```

```python
df.head(3)
```

```python
df.columns
```

```python
#Eliminamos las comillas de todas las celdas.
df['reportedtimestamp,eventdate,eventtime,shape,duration,witnesses,weather,firstname,lastname,latitude,longitude,sighting,physicalevidence,contact,researchoutcome']=df['reportedtimestamp,eventdate,eventtime,shape,duration,witnesses,weather,firstname,lastname,latitude,longitude,sighting,physicalevidence,contact,researchoutcome'].apply(lambda x: x.replace('"',"") if type (x) == str else str (x))
```

```python
#Separamos la única columna que tenemos hasta ahora en varias columnas.
df[['reported_time_stamp','event_date','event_time','shape','duration','witnesses','weather','first_name','last_name','latitude','longitude','sighting','physical_evidence','contact','research_outcome']] = df['reportedtimestamp,eventdate,eventtime,shape,duration,witnesses,weather,firstname,lastname,latitude,longitude,sighting,physicalevidence,contact,researchoutcome'].str.split(",",expand = True)
df.head(2)
```

```python
df[['reported_time_stamp_date','reported_time_stamp_time']] = df['reported_time_stamp'].str.split("T",expand = True)
df.head(3)
```

```python
#Eliminamos las columna donde hemos hecho el split
df.drop(['reportedtimestamp,eventdate,eventtime,shape,duration,witnesses,weather,firstname,lastname,latitude,longitude,sighting,physicalevidence,contact,researchoutcome'],axis =1, inplace=True)
df.head()
```

```python
df[['event_time_hour','event_time_minutes']] = df['event_time'].str.split(":",expand = True)
df.head(3)
```

```python
df['name']=df['first_name'].str.cat(df['last_name'],sep=' ')
df.head(3)
```

```python
df[['reported_time_stamp_time_hour','reported_time_stamp_time_min','reported_time_stamp_time_rest']] = df['reported_time_stamp_time'].str.split(":",expand = True)
df.head(3)
```

```python
df['reported_time_stamp_time']=df['reported_time_stamp_time_hour'].str.cat(df['reported_time_stamp_time_min'],sep=':')
df.head(3)
```

```python
#Eliminamos las columnas que nos nos interesan ya.
df.drop(['reported_time_stamp','event_time_minutes','reported_time_stamp_time_min','reported_time_stamp_time_rest'],axis =1, inplace=True)
df.head(3)
```

```python
nuevo_orden=['reported_time_stamp_date','reported_time_stamp_time','reported_time_stamp_time_hour','event_date','event_time','event_time_hour','shape','duration','witnesses',
                'weather','first_name','last_name','name','latitude','longitude','sighting','physical_evidence','contact','research_outcome']
df=df.reindex(columns=nuevo_orden)
df.head(3)
```

```python
#Vemos que la columna sighting no es necesaria porque es Y es todas las filas.
df['sighting'].value_counts()
```

```python
df.drop(['sighting'],axis =1, inplace=True)
df.head(3)
```

```python
#Vemos que en esta columna tenemos el valor 'probable' y 'unexplained' que podrían ser lo mismo, ya que ambas indican la posible presencia de UFOs.
df['research_outcome'].value_counts()
```

```python
df['research_outcome']=np.where(df['research_outcome']=='explained','explained','unexplained')
```

```python
df['research_outcome'].value_counts()
```

```python
#Vemos que no tenemos valores repetidos en nuestro dataframe.
df.duplicated().sum()
```

```python
df['reported_time_stamp_time_hour']=df['reported_time_stamp_time_hour'].astype(int)
```

```python
df['event_time_hour']=df['event_time_hour'].astype(int)
```

```python
df.info()
```

```python
condiciones=[(df['reported_time_stamp_time_hour']>=22) & (df['reported_time_stamp_time_hour']<24),
             (df['reported_time_stamp_time_hour']>=0) & (df['reported_time_stamp_time_hour']<6),
             (df['reported_time_stamp_time_hour']>=6) & (df['reported_time_stamp_time_hour']<9),
             (df['reported_time_stamp_time_hour']>=9) & (df['reported_time_stamp_time_hour']<18),
             (df['reported_time_stamp_time_hour']>=18) & (df['reported_time_stamp_time_hour']<22)]
opciones=['night','night','sunrise','day','sunset']
df['reported_time_stamp_daytime']=np.select(condiciones,opciones)
```

```python
df.head(3)
```

```python
condiciones2=[(df['event_time_hour']>=22) & (df['event_time_hour']<24),
             (df['event_time_hour']>=0) & (df['event_time_hour']<6),
             (df['event_time_hour']>=6) & (df['event_time_hour']<9),
             (df['event_time_hour']>=9) & (df['event_time_hour']<18),
             (df['event_time_hour']>=18) & (df['event_time_hour']<22)]
opciones2=['night','night','sunrise','day','sunset']
df['event_time_daytime']=np.select(condiciones2,opciones2)
```
